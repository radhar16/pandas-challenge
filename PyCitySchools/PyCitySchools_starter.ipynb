```python
# Dependencies and Setup
import pandas as pd

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas DataFrames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset.  
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

## District Summary

* Calculate the total number of schools

* Calculate the total number of students

* Calculate the total budget

* Calculate the average math score 

* Calculate the average reading score

* Calculate the percentage of students with a passing math score (70 or greater)

* Calculate the percentage of students with a passing reading score (70 or greater)

* Calculate the percentage of students who passed math **and** reading (% Overall Passing)

* Create a dataframe to hold the above results

* Optional: give the displayed data cleaner formatting

```python
#To get the distinct school name by using nunique() method
Total_Schools = school_data_complete["school_name"].nunique()

# to get the total students by using length of column student name
Total_Students = len(school_data_complete["student_name"])

#to get the total budget of school from school data csv file by using sum() method
Total_Budget = school_data["budget"].sum()

#to get the average math score of students by using mean() method
Avg_math_score = school_data_complete["math_score"].mean()

#to get the average reading score of students by using mean() method
Avg_read_score = school_data_complete["reading_score"].mean()

#to get the % of students passing math by using loc() method
passing_math_70 = school_data_complete.loc[school_data_complete["math_score"] >=70,:]
passing_math = (len(passing_math_70["math_score"])/len(school_data_complete["math_score"]))*100

#to get the % of students passing reading by using loc() method
passing_read_70 = school_data_complete.loc[school_data_complete["reading_score"] >=70,:]
passing_read = (len(passing_read_70["reading_score"])/len(school_data_complete["reading_score"]))*100

#to get the % of students passing overall by using loc() method
overall_pass_70 =  school_data_complete.loc[(school_data_complete["math_score"] >=70) & (school_data_complete["reading_score"] >=70),:]
overall_pass = (len(overall_pass_70["Student ID"])/len(school_data_complete["Student ID"]))*100

#creating a dataframe to hold the summary
Dist_Summary = pd.DataFrame ({'Total Schools': Total_Schools,
                                'Total Students': Total_Students,
                                'Total Budget': Total_Budget,
                                'Average Math Score': Avg_math_score,
                                'Average Reading Score': Avg_read_score,
                                 '% Passing Math': passing_math,
                                 '% Passing Reading':passing_read,
                                 '% Overall Passing':overall_pass
                                }, index=[0])

#Formatting columns to display cleaner data
Dist_Summary["Total Students"] =  Dist_Summary["Total Students"].map("{:,}".format)
Dist_Summary["Total Budget"] = Dist_Summary["Total Budget"].map("${:,.2f}".format)
Dist_Summary["Average Math Score"] = Dist_Summary["Average Math Score"].map("{:.2f}".format)
Dist_Summary["Average Reading Score"] = Dist_Summary["Average Reading Score"].map("{:.2f}".format)
Dist_Summary["% Passing Math"] = Dist_Summary["% Passing Math"].map("{:.2f}".format)
Dist_Summary["% Passing Reading"] = Dist_Summary["% Passing Reading"].map("{:.2f}".format)
Dist_Summary["% Overall Passing"] = Dist_Summary["% Overall Passing"].map("{:.2f}".format)

Dist_Summary
```

## School Summary

* Create an overview table that summarizes key metrics about each school, including:
  * School Name
  * School Type
  * Total Students
  * Total School Budget
  * Per Student Budget
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * % Overall Passing (The percentage of students that passed math **and** reading.)
  
* Create a dataframe to hold the above results

```python
#create a groupby dataframe by school name and also setting the school name as index
group_schoolname = school_data_complete.set_index("school_name").groupby(["school_name"])

#to create a school summary table, get the school type from school data table and group it on school name
school_type = school_data.set_index("school_name")["type"]

#need this school size series for the segment Scores by School Size
school_size = school_data.set_index("school_name")["size"]

#to get the total students counts by using count() method
group_total_students = group_schoolname["Student ID"].count()

#get the school budget from school data table and group it on school name
group_total_budget = school_data.set_index("school_name")["budget"]

#calculating the per student budget
per_stu_budget = group_total_budget/group_total_students

#calculating the average math score
avg_math_score = group_schoolname["math_score"].mean()

#calculating the average reading score
avg_read_score = group_schoolname["reading_score"].mean()

#calculating the % passing math by using loc() method from merged school and student dataframes
grp_passing_math = (school_data_complete.loc[school_data_complete["math_score"] >=70].groupby(["school_name"])["Student ID"].count()/group_total_students)*100

#calculating the % passing reading by using loc() method from merged school and student dataframes
grp_passing_read = (school_data_complete.loc[school_data_complete["reading_score"] >=70].groupby(["school_name"])["Student ID"].count()/group_total_students)*100

#calculating the % overall passing by using loc() method from merged school and student dataframes
overall_passing = (school_data_complete.loc[(school_data_complete["math_score"] >=70) & (school_data_complete["reading_score"] >=70)].groupby(["school_name"])["Student ID"].count()/group_total_students)*100

# creating a dataframe to hold the results
school_summary = pd.DataFrame({"Size": school_size,
                               "School Type": school_type,
                               "Total Students": group_total_students,
                               "Total School Budget": group_total_budget,
                               "Per Student Budget": per_stu_budget,
                               "Average Math Score": avg_math_score,
                               "Average Reading Score": avg_read_score,
                               "% Passing Math": grp_passing_math,
                               "% Passing Reading": grp_passing_read,
                               "% Overall Passing": overall_passing})

#Creating dataframe to hold data and display the results formatted
formatted_school_summary = school_summary.loc[:, ['School Type','Total Students','Total School Budget','Per Student Budget',
                                                 'Average Math Score','Average Reading Score','% Passing Math','% Passing Reading',
                                                 '% Overall Passing']]

# #Formatting columns to display cleaner data
formatted_school_summary["Total Students"] = formatted_school_summary["Total Students"].map("{:,}".format)
formatted_school_summary["Total School Budget"] = formatted_school_summary["Total School Budget"].map("${:,.2f}".format)
formatted_school_summary["Per Student Budget"] =  formatted_school_summary["Per Student Budget"].map("${:.2f}".format)
formatted_school_summary["Average Math Score"] = formatted_school_summary["Average Math Score"].map("{:.2f}".format)
formatted_school_summary["Average Reading Score"] = formatted_school_summary["Average Reading Score"].map("{:.2f}".format)
formatted_school_summary["% Passing Math"] = formatted_school_summary["% Passing Math"].map("{:.2f}".format)
formatted_school_summary["% Passing Reading"] = formatted_school_summary["% Passing Reading"].map("{:.2f}".format)
formatted_school_summary["% Overall Passing"] = formatted_school_summary["% Overall Passing"].map("{:.2f}".format)

formatted_school_summary

```

## Scores by School Spending

* Create a table that breaks down school performances based on average Spending Ranges (Per Student). Use 4 reasonable bins to group school spending. Include in the table each of the following:
  * Average Math Score
  * Average Reading Score
  * % Passing Math
  * % Passing Reading
  * Overall Passing Rate (Average of the above two)

```python
#creating a new data frame to hold only required columns for school spending summary from school_summary table
school_spend_sel_col = school_summary.loc[:,["Per Student Budget","Average Math Score", "Average Reading Score","% Passing Math",
                          "% Passing Reading", "% Overall Passing"]]

# Creating the bins in which Data will be held
bins = [0, 585, 630, 645, 680]

# Creating the names for the four bins
group_names = ["<$585", "$585-630", "$630-645", "$645-680"]

#adding a column to hold the group names for the bins
school_spend_sel_col["Spending Ranges (Per Student)"] = pd.cut(school_spend_sel_col["Per Student Budget"], bins, labels=group_names, include_lowest=True)

#create a groupby dataframe to group school spending
group_spend_range = school_spend_sel_col.set_index("Spending Ranges (Per Student)").groupby(["Spending Ranges (Per Student)"]).mean()

#creating a new data frame to hold only required columns and display spend summary
summary_spend = group_spend_range[["Average Math Score", "Average Reading Score","% Passing Math",
                          "% Passing Reading", "% Overall Passing"]]

#create dataframe to hold the formatting columns 
formatted_summary_spend = pd.DataFrame(summary_spend)

#Formatting the columns to display cleaner data
formatted_summary_spend["Average Math Score"] = formatted_summary_spend["Average Math Score"].map("{:.2f}".format)
formatted_summary_spend["Average Reading Score"] =  formatted_summary_spend["Average Reading Score"].map("{:.2f}".format)
formatted_summary_spend["% Passing Math"] =  formatted_summary_spend["% Passing Math"].map("{:.2f}".format)
formatted_summary_spend["% Passing Reading"] =  formatted_summary_spend["% Passing Reading"].map("{:.2f}".format)
formatted_summary_spend["% Overall Passing"] =  formatted_summary_spend["% Overall Passing"].map("{:.2f}".format)

formatted_summary_spend

```
